```python
N = int(input())
num = list(map(int, input().split()))

num.sort()
cnt = 0
result = 0

for n in num:
    cnt+=1
    if n <= cnt:
        cnt = 0
        result += 1
print(result)
```

### 03  


```python
num = input()
result=0

for i in range(1, len(num)-1):
    if num[i] != num[i+1] and num[i+1] != num[0]:
        result+=1
print(result)
```
