## Q1 백준 1072 번

```python
x, y = map(int, input().split())
z = (y * 100) // x

# 승률이 100%일 때는 더 이상 승률을 변화시킬 수 없으므로 -1 출력
if z >= 99:
    print(-1)
else:
    # 이분 탐색으로 최소 게임 수 찾기
    left, right = 0, 1000000000  # 게임 횟수의 최대 범위 설정
    while left < right:
        mid = (left + right) // 2
        new_z = ((y + mid) * 100) // (x + mid)
        if new_z > z:
            right = mid
        else:
            left = mid + 1
    print(left)
```

```python
arr = input().split()
x = int(arr[0])
y = int(arr[1])
target_z = (y*100) / 100
# z는 걍 내가 임의로 검사해볼 값
count = 0
def targeting(x,y,tar_z,count):
  z = (y+1)*100/(x+1)
  if z == tar_z :
    count += 1
    return targeting(x+2,y+2,tar_z,count) # 루프 한번 더 돌고
  else:
    return count # 최종으론 count 개수를 반환

if targeting(x,y,target_z,count) == 0:
  print(-1)
else :
  print(count)
```

## 백준 23326 번


```python
from sys import stdin, stdout

input = stdin.read
def solve():
    data = input().split()
    N = int(data[0])
    Q = int(data[1])
    A = list(map(int, data[2:2+N]))
    queries = data[2+N:]

    pos = 0  # 도현이의 현재 위치
    sights = set()  # 명소 위치를 저장하는 set

    # 초기 명소들을 관리
    for i in range(N):
        if A[i] == 1:
            sights.add(i)

    output = []
    idx = 0
    for q in queries:
        if q == '1':
            i = int(queries[idx + 1]) - 1
            idx += 2
            if A[i] == 0:
                A[i] = 1
                sights.add(i)
            else:
                A[i] = 0
                sights.remove(i)
        elif q == '2':
            x = int(queries[idx + 1])
            idx += 2
            pos = (pos + x) % N
        elif q == '3':
            idx += 1
            if not sights:
                output.append("-1")
            else:
                # 최소 이동 거리 찾기
                min_moves = float('inf')
                for sight in sights:
                    # 시계 방향으로 움직인 최소 거리 계산
                    moves = (sight - pos) % N
                    if moves < min_moves:
                        min_moves = moves
                output.append(str(min_moves))

    stdout.write("\n".join(output) + "\n")

solve()
```

## 백준 2470번

```python
# 입력 받기
N = int(input())
arr = list(map(int, input().split()))

# 배열을 정렬
arr.sort()

# 두 포인터 초기화
left = 0
right = N - 1
closest_sum = float('inf')  # 가능한 가장 큰 값으로 초기화
best_pair = (0, 0)

# 두 포인터가 교차할 때까지 반복
while left < right:
    current_sum = arr[left] + arr[right]

    # 현재 합의 절대값이 이전의 가장 작은 절대값보다 작다면 정보 업데이트
    if abs(current_sum) < abs(closest_sum):
        closest_sum = current_sum
        best_pair = (arr[left], arr[right])

    # 합을 조절하여 0에 더 가까워지도록 포인터 이동
    if current_sum < 0:
        left += 1  # 합을 증가시키기 위해 왼쪽 포인터 이동
    else:
        right -= 1  # 합을 감소시키기 위해 오른쪽 포인터 이동

# 결과 출력
print(best_pair[0], best_pair[1])
```
