```python
N = int(input())
cnt = 0

money = [500,100,50,10]

for m in money:    
    cnt += N // m
    N = N % m
print(cnt)
```

```python
N,M,K = map(int, input().split())
num = list(map(int, input().split()))
num.sort(reverse=True)

total = 0
n_times = M // (K+1)
total += num[0]*n_times*K
total += num[1]*n_times
total += num[0]*(M%(K+1))
print(total)
```

```python
N,M = map(int,input().split())
result = 0

while N!=0:
    num = list(map(int, input().split()))
    min = min(num)
    result = max(result, min)
    N -= 1
print(result)
```

```python
N,K = map(int,input().split())
cnt = 0

while N != 1:
    if N % K == 0:
        N = N // K
        cnt += 1
    else:
        N -= 1
        cnt += 1
print(cnt)
```
